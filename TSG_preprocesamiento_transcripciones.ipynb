```python
import os
import pandas as pd
todas_lines = []
folder_path = './transcripts/8000s_duplicate/8000_a/markdown_converted_via_word_processed/'
for filename in os.listdir(folder_path):
    if filename.endswith('.md'):
        with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
            content = file.readlines()[9:]
            
            data = [(filename,line[:2].strip(),line[2:].strip()) for line in content if line.strip()]
            if any(not (line.startswith('I:') or line.startswith('S:') or line.startswith('@')) for line in content if line.strip()):
                print(f"El archivo {filename} contiene líneas que no empiezan por I:, S: o @.")
                ejemplo = next((line for line in content if not (line.startswith('I:') or line.startswith('S:') or line.startswith('@')) and line.strip()), None)
                if ejemplo:
                    print(ejemplo[:2].strip(),'||||',ejemplo[2:].strip())
                    print("              Ejemplo de línea:", ejemplo.strip())
                
            todas_lines.extend(data)
            
```

```python
lines = pd.DataFrame(todas_lines, columns=['name', 'prefix', 'rest'])
```

```python
import os

def cargar_archivos_arreglados(file_list, folder_path):
    archivos_arreglados = []
    for filename in file_list:
        if filename.endswith('.md'):
            with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
                content = file.readlines()[2:]
                intervencion = ""
                intervencion_prefix = ""
                for line in content:

                    line = line.strip()

                    if not line:
                        continue
                    if line.startswith(('S:', 'I:', '@')):
                        if intervencion:
                            print(line)
                            archivos_arreglados.append((filename, intervencion_prefix, intervencion))
                            intervencion = ""
                            intervencion_prefix = ""
                        # archivos_arreglados.append((filename, line[:2].strip(), line[2:].strip()))
                        intervencion_prefix = line[:2].strip()
                        intervencion = line[2:].strip()
                    else:
                        # Si la línea no empieza por S:, I: o @, se asume que es continuación de la intervención del paciente
                        intervencion += " " + line
            if intervencion:
                archivos_arreglados.append((filename, intervencion_prefix, intervencion))
    return archivos_arreglados

# Ejemplo de uso:
file_list = ['80270_fu1_tran.md', '80270_fu2_tran.md']
archivos_arreglados = cargar_archivos_arreglados(file_list, folder_path)
```

```python
#añadimos las filas arregladas al df lines
lines = pd.concat([lines, pd.DataFrame(archivos_arreglados, columns=['name', 'prefix', 'rest'])], ignore_index=True)
```
